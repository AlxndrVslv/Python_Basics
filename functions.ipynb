```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
import time
```

## Задание 1

```python
def doc_name(data):       
    '''
    Выводит владельца документа по номеру
    '''
    for doc in documents:
        if doc['number'] == data:
            return doc['name']
    return None        

def dir_name(data):
    '''
    Выводит номер полки по номеру документа
    '''
    for dir, docs in directories.items():
        if data in docs: 
            return dir
    return None 

def doc_info(cmd, data):
    '''
    Выводит результат
    '''
    res = 'Документ не найден в базе'
    if cmd == 'p':
        dn = doc_name(data)
        if dn: res = f"Владелец документа: {dn}"
    elif cmd == 's':
        dn = dir_name(data)
        if dn: res = f"Документ хранится на полке: {dn}"
    print(res)
    
def dir_keys_list():
    '''
    Формирует список полок
    '''
    return ', '.join(list(directories.keys()))

def dir_modify(cmd, data):
    '''
    Вставка/удаление полок
    '''
    if cmd == 'ads':
        if data not in directories.keys():
            directories.setdefault(data, [])
            print(f"Полка добавлена. Текущий перечень полок: {dir_keys_list()}")
        else:
            print(f"Такая полка уже существует. Текущий перечень полок: {dir_keys_list()}")
    if cmd == 'ds':
        if data in directories.keys():
            if directories[data] == []:
                directories.pop(data)
                print(f"Полка удалена. Текущий перечень полок: {dir_keys_list()}")
            else:
                print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {dir_keys_list()}")
        else:
            print(f"Такой полки не существует. Текущий перечень полок: {dir_keys_list()}")

def doc_list():
    '''
    Вывод всего списка
    '''
    for doc in documents:
        res = f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}"
        for dir, docs in directories.items():
            if doc['number'] in docs:
                res += f', полка хранения: {dir}'
        print(res)

def main(cmd):
    '''
    Заполнение параметров команды
    '''
    if cmd in ('p', 's'):
        return doc_info(cmd, input('Введите номер документа'))
    elif cmd == 'l':
        return doc_list()
    elif cmd in ('ads', 'ds'):
        return dir_modify(cmd, input('Введите номер полки'))
    else:
        print(f"Не известная команда '{cmd}'")

cmd = input('Введите команду:')
while cmd != 'q':
    main(cmd)
    time.sleep(0.5) # Задержка, чтобы сообщение успело быть выведенным на экран до новой команды ввода
    cmd = input('Введите команду:')
print('Выход')
```

## Задание 2

```python
########################################################

def doc_name(data):       
    '''
    Выводит владельца документа по номеру
    '''
    for doc in documents:
        if doc['number'] == data:
            return doc['name']
    return None        

########################################################

def dir_name(data):
    '''
    Выводит номер полки по номеру документа
    '''
    for dir, docs in directories.items():
        if data in docs: 
            return dir
    return None 

########################################################

def doc_info(cmd, data):
    '''
    Выводит результат
    '''
    res = 'Документ не найден в базе'
    if cmd == 'p':
        dn = doc_name(data)
        if dn: res = f"Владелец документа: {dn}"
    elif cmd == 's':
        dn = dir_name(data)
        if dn: res = f"Документ хранится на полке: {dn}"
    print(res)
    
########################################################

def dir_keys_list():
    '''
    Формирует список полок
    '''
    return ', '.join(list(directories.keys()))

########################################################

def dir_modify(cmd, data):
    '''
    Вставка/удаление полок
    '''
    if cmd == 'ads':
        if data not in directories.keys():
            directories.setdefault(data, [])
            print(f"Полка добавлена. Текущий перечень полок: {dir_keys_list()}")
        else:
            print(f"Такая полка уже существует. Текущий перечень полок: {dir_keys_list()}")
    if cmd == 'ds':
        if data in directories.keys():
            if directories[data] == []:
                directories.pop(data)
                print(f"Полка удалена. Текущий перечень полок: {dir_keys_list()}")
            else:
                print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {dir_keys_list()}")
        else:
            print(f"Такой полки не существует. Текущий перечень полок: {dir_keys_list()}")

########################################################

def doc_list():
    '''
    Вывод всего списка
    '''
    for doc in documents:
        res = f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}"
        for dir, docs in directories.items():
            if doc['number'] in docs:
                res += f', полка хранения: {dir}'
        print(res)

########################################################

def add_doc(doc_num, dir, 
            doc_add = True, 
            doc_type = None, doc_holder = None
            ):
    '''
    Добавление нового документа
    doc_add: 
    True - добавляем документ в documents и directories; 
    False - только в directories
    '''
    if dir in directories.keys():
        if doc_add: 
            documents.append({'type': doc_type, 'number': doc_num, 'name': doc_holder})
        directories[dir].append(doc_num)
        return True
    else:
        return False
    

########################################################

def del_doc(doc_num, doc_remove = True):
    '''
    Удаление документа
    doc_remove: 
    True - удаляем документ из documents и directories; 
    False - только из directories
    '''        
    found = False
    documents_copy = documents.copy()
    if doc_remove:
        for doc in documents_copy:
            if doc['number'] == doc_num:
                found = True
                documents.remove(doc)                        
    
    for dir_docs in directories.values():
        if doc_num in dir_docs:
            found = True
            dir_docs.remove(doc_num)

    if found: return True
    else: return False            

########################################################

def move_doc(doc_num, dir):
    '''
    Перемещение документа
    '''
    doc_exist = doc_num in set(sum(directories.values(), []))
    dir_exist = dir in directories.keys()

    if doc_exist and dir_exist:
        del_doc(doc_num, doc_remove = False)
        add_doc(doc_num, dir, doc_add = False)
    
    return (doc_exist, dir_exist)

########################################################

def main(cmd):
    '''
    Заполнение параметров команды
    '''
    if cmd in ('p', 's'):
        return doc_info(cmd, input('Введите номер документа'))
    elif cmd == 'l':
        return doc_list()
    elif cmd in ('ads', 'ds'):
        return dir_modify(cmd, input('Введите номер полки'))
    elif cmd == 'ad':
        if  add_doc(    doc_num =       input('Введите номер документа'),
                        doc_type =      input('Введите тип документа'),
                        doc_holder =    input('Введите владельца документа'),
                        dir =           input('Введите полку хранения')
                   ): print('Документ добавлен. Текущий список документов:')
        else: print("Такой полки не существует. Добавьте полку командой 'ads'.")
        doc_list();            
    elif cmd == 'd':
        if del_doc(doc_num = input('Введите номер документа')):
            print('Документ удален.')
        else:
            print('Документ не найден в базе.') 
        print('Текущий список документов:')
        doc_list();
    elif cmd == 'm':        
        res = move_doc  (   doc_num =   input('Введите номер документа'),
                            dir =       input('Введите полку хранения')
                        )
        if res == (True, False):            
            print(f"Такой полки не существует. Текущий перечень полок: {dir_keys_list()}")
        else:
            if res == (True, True):
                print('Документ перемещен')
            if res == (False, True):
                print('Документ не найден в базе.')
            if res == (False, False):
                print('Документ не найден в базе. Полка не существует')
            print('Текущий список документов:')
            doc_list();        
    else:
        print(f"Не известная команда '{cmd}'")

########################################################

cmd = input('Введите команду:')
while cmd != 'q':
    main(cmd)
    time.sleep(0.5) # Задержка, чтобы сообщение успело быть выведенным на экран до новой команды ввода
    cmd = input('Введите команду:')
print('Выход')
```
